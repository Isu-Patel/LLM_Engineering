```python
import os
from dotenv import load_dotenv

load_dotenv(override=True)
api_key = os.getenv('OPENAI_KEY')

if not api_key:
    print("No API key was fouund - please head over to the troubleshooting notebook in this folder to identify & fix the error.")
elif not api_key.startswith("sk-proj-"):
    print("An API key was found, but it doesn't appear to start with sk-proj-. Please check for the key if it is correct or not.")
else:
    print("API key was found and looks good.")
```

```python
import requests

header = {"Authorization": f"Bearer {api_key}", "Content-Type": "application/json"}

payload = {
    "model": "gpt-5-nano",
    "messages": [{"role": "user", "content": "Tell me a Funny Joke."}]
}

payload
```

```python
response = requests.post(
    "https://api.openai.com/v1/chat/completions",
    headers=header,
    json=payload
)

response.json()
```

```python
from openai import OpenAI
openai = OpenAI(api_key=api_key)

response = openai.chat.completions.create(model="gpt-5-nano", messages=[{"role": "user", "content": "Tell me a Funny Joke."}])

response.choices[0].message.content
```
